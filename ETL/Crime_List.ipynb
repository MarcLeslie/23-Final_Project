```python
# When ypu specify columns, all other columns are dropped 
crime_grouped_df = crime_list_df.groupby(['Year', 'LAPD_Area', "Crime_Type"]).agg({'Crime_Type':['count']})
```

```python
crime_grouped_df.head()
```

```python
crime_grouped_df.columns = ['Crime_Counts'] 
crime_grouped_df.reset_index(level=['Year','LAPD_Area',"Crime_Type"],inplace=True)
```

```python
crime_grouped_df.head(100)
```

```python
crime_grouped_df['Crime_Type'] = crime_grouped_df['Crime_Type'].str.replace('Rape/Sexual Assault', 'Rape/SA')
```

```python
crime_grouped_df.head()
```

```python
# OUTPUT TO JSON, LOAD TO GITHUB, GET RID OF THE INDEX THAT IS MESSING UP THE JSON
crime_grouped_df.to_json(r'../static/data/crime_grouped.json', orient = "records")
```
